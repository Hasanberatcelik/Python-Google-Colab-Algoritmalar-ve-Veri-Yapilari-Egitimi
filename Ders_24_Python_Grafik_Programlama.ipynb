# Bölüm 24: Python Programlama Dili Üzerinde Grafik Programlama

# 24.1 Grafik Programlama Nedir?
Grafik programlama, Python gibi programlama dillerinde çizim yapma, görsel nesne oluşturma, arayüz geliştirme ve etkileşimli uygulamalar hazırlama sürecidir. Genellikle görsel çıktı gerektiren oyunlar, arayüzler, veri görselleştirmeleri ve simülasyonlar için kullanılır.

# 24.2 Python'da Kullanılan Popüler Grafik Kütüphaneleri

| Kütüphane     | Kullanım Amacı                             | Uygulama Alanı                    |
|---------------|---------------------------------------------|------------------------------------|
| `turtle`      | Temel çizim ve algoritma öğretimi           | Eğitim, çocuklar için kodlama      |
| `matplotlib`  | Grafik çizimi ve veri görselleştirme        | Bilimsel grafikler, veri analizi   |
| `pygame`      | Oyun ve görsel simülasyon geliştirme        | 2D oyun motoru, interaktif uygulama|
| `Tkinter`     | GUI (Grafiksel Arayüz) oluşturma            | Formlar, butonlar, menüler         |
| `OpenCV`      | Görüntü işleme ve kamera uygulamaları       | Bilgisayarla görü, AI uygulamaları |
| `Plotly`      | Etkileşimli web tabanlı grafikler           | Web dashboard, veri sunumları      |
| `PyQt5`       | Profesyonel masaüstü arayüzler              | Ticari uygulamalar, yazılım araçları|

```python
import turtle
from pyvirtualdisplay import Display # Display sınıfını içe aktarıyoruz
import matplotlib.pyplot as plt
import numpy as np

# Başlangıç konumu ve açı
x, y = [0], [0]
açı = 0  # İlk yön sağa (0 derece)

# Kalemin hareketlerini tanımla (turtle.forward ve turtle.right(90) karşılığı)
uzunluk = 100

# 4 kenarlı bir şekil (örneğin kare) çizmek için döngü
for i in range(4):
    # Son konuma göre ileri doğru yeni nokta hesaplanır
    x_son = x[-1] + uzunluk * np.cos(np.radians(açı))
    y_son = y[-1] + uzunluk * np.sin(np.radians(açı))

    # Hesaplanan yeni noktalar listeye eklenir
    x.append(x_son)
    y.append(y_son)

    # Sağa doğru 90 derece dönüş yapılır
    açı -= 90

# Grafik çizimi
plt.figure(figsize=(5,5))
plt.plot(x, y, marker='o', color='blue')
plt.title("Matplotlib ile Turtle Kare Çizimi")
plt.axis('equal')
plt.grid(True)
plt.show()
```

```python
# Matplotlib ile Grafik Çizimi

import matplotlib.pyplot as plt

x = [1, 2, 3, 4]
y = [2, 4, 6, 8]

plt.plot(x, y, marker='o')
plt.title("Doğrusal Grafik")
plt.xlabel("X Ekseni")
plt.ylabel("Y Ekseni")
plt.grid(True)
plt.show()
```

```python
import matplotlib.pyplot as plt

# Ekran 400x300 boyutlarında tanımlanmıştı
genislik, yukseklik = 400, 300

# Daire bilgileri
merkez_x, merkez_y = 200, 150
yaricap = 50

# Matplotlib figürü oluştur
fig, ax = plt.subplots(figsize=(6, 4.5))
ax.set_xlim(0, genislik)
ax.set_ylim(0, yukseklik)

# Y eksenini ters çevirmek (Koordinat sisteminde)
ax.invert_yaxis()

# Daireyi çiz
daire = plt.Circle((merkez_x, merkez_y), yaricap, color='red')
ax.add_patch(daire)

# Arka plan beyaz olsun
fig.patch.set_facecolor('white')
ax.set_facecolor('white')

# Eksenleri kapat
ax.axis('off')
plt.title("Matplotlib ile Grafik Uygulaması", fontsize=14)
plt.show()
```
